```python
#Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read CSV as pandas data frame
import pandas as pd
df1=pd.read_csv('/content/drive/MyDrive/CompAndViz/Education.csv')
df2=pd.read_csv('/content/drive/MyDrive/CompAndViz/Housing.csv')
df3=pd.read_csv('/content/drive/MyDrive/CompAndViz/Population.csv')
df4=pd.read_csv('/content/drive/MyDrive/CompAndViz/Poverty.csv')
df5=pd.read_csv('/content/drive/MyDrive/CompAndViz/Median income.csv')
```

```python
df1.rename(columns={'aphic Area Name': 'Zip code'}, inplace=True)
df2.rename(columns={'aphic Area Name': 'Zip code'}, inplace=True)
df3.rename(columns={'aphic Area Name': 'Zip code'}, inplace=True)
df4.rename(columns={'aphic Area Name': 'Zip code'}, inplace=True)
df5.rename(columns={'aphic Area Name': 'Zip code'}, inplace=True)
```

```python
df1
```

```python
df1 = df1.drop(columns=['Geographic Area Name'])
df2 = df2.drop(columns=['Geographic Area Name'])
df3 = df3.drop(columns=['Geographic Area Name'])
df4 = df4.drop(columns=['Geographic Area Name'])
df5 = df5.drop(columns=['Geographic Area Name'])
```

```python
merged_df = df1.merge(df2, on='Zip code') \
                .merge(df3, on='Zip code')\
                .merge(df4, on='Zip code')\
                .merge(df5, on='Zip code')

print(merged_df)
```

```python
merged_df
```

```python
merged_df[' Median House value (dollars)'] = merged_df[' Median House value (dollars)'].replace(r'\+$', '', regex=True)
```

```python
merged_df
```

```python
merged_df[' Median House value (dollars)'] = merged_df[' Median House value (dollars)'].replace({',': ''}, regex=True).astype(int)
```

```python
merged_df.dtypes
```

```python
print(merged_df.columns)
```

```python
# Normalize the columns
merged_df['Education_norm'] = (merged_df["Graduate or professional school"] - merged_df["Graduate or professional school"].min()) / (merged_df["Graduate or professional school"].max() - merged_df["Graduate or professional school"].min())
merged_df['Population_norm'] = (merged_df['Population'] - merged_df['Population'].min()) / (merged_df['Population'].max() - merged_df['Population'].min())
merged_df['Poverty_norm'] = (merged_df['Population for whom poverty status is determined'] - merged_df['Population for whom poverty status is determined'].min()) / (merged_df['Population for whom poverty status is determined'].max() - merged_df['Population for whom poverty status is determined'].min())
merged_df['Income_norm'] = (merged_df[' Median family income (4ppl)'] - merged_df[' Median family income (4ppl)'].min()) / (merged_df[' Median family income (4ppl)'].max() - merged_df[' Median family income (4ppl)'].min())
merged_df['Housing_norm'] = (merged_df[' Median House value (dollars)'] - merged_df[' Median House value (dollars)'].min()) / (merged_df[' Median House value (dollars)'].max() - merged_df[' Median House value (dollars)'].min())

# Adjust Poverty_norm for negative correlation (higher poverty -> lower score)
merged_df['Poverty_norm'] = 1 - merged_df['Poverty_norm']

# Assign weights
weights = {
    'Education_norm': 0.3,   # Weight for education level
    'Population_norm': 0.25,   # Weight for population
    'Poverty_norm': 0.2,      # Weight for poverty
    'Income_norm': 0.2,       # Weight for income
    'Housing_norm': 0.15      # Weight for housing value
}

# Calculate the composite score
merged_df['Score'] = (
    merged_df['Education_norm'] * weights['Education_norm'] +
    merged_df['Population_norm'] * weights['Population_norm'] +
    merged_df['Poverty_norm'] * weights['Poverty_norm'] +
    merged_df['Income_norm'] * weights['Income_norm'] +
    merged_df['Housing_norm'] * weights['Housing_norm']
)

# Rank the zip codes based on the score
merged_df['Rank'] = merged_df['Score'].rank(ascending=False).astype(int)

new_df = merged_df[['Zip code', "Graduate or professional school", 'Population', 'Population for whom poverty status is determined', ' Median family income (4ppl)', ' Median House value (dollars)', 'Score', 'Rank']]

# Display the updated DataFrame
print(merged_df[['Zip code', 'Score', 'Rank']])
```
